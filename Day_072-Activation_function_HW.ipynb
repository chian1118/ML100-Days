# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return np.maximum(0, x)

def dReLU(x):
    return (1 * (x > 0))
```

```python
x = plt.linspace(-10,10,100)
```

```python
plt.plot(x, ReLU(x), color='red')
plt.title('ReLU')
```

```python
plt.plot(x, dReLU(x), color='b')
plt.title('dReLU')
```
