### 線性回歸模型能夠準確預測非線性關係的資料集嗎?

Ans:不行

### 回歸模型是否對資料分布有基本假設?
Ans:沒有基本假設，但回歸模型背後有其基本假設
1. 殘差為常態
2. 殘差是否有相同的分布，即變異數相等
3. 殘差是否獨立
4. 線性關係是否正確
